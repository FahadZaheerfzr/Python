This code shows how to detect faces using OpenCV library.

```
# Importing the library
import cv2
```

```
# Adding the paths of image and haar cascade
imagePath = "/content/Test.jpg"
cascPath = "/content/haarcascade_frontalface_default.xml"
```

```
# Creating the haar cascade
faceCascade = cv2.CascadeClassifier(cascPath)
```

```
# Read the image
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```
# Detect faces in the image
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30),
    flags = cv2.CASCADE_SCALE_IMAGE
)
```

```
# Print the number of faces found
print("Found {0} faces!".format(len(faces)))
# Draw a rectangle around the faces
for (x, y, w, h) in faces:
  cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```
